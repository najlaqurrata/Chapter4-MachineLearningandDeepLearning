## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Tenure', 'Balance','NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']].copy()
df
```

```python
df.info()
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df, columns=['Geography', 'Gender'])
```

```python
df
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.drop("Exited", axis=1)
y = df["Exited"]
```

```python
X.head()
```

```python
y.head()
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
#X_transform = X.copy()
```

```python
X_transform
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

Model pertama yang digunakan adalah Decision Tree. Decision Tree adalah model yang digunakan untuk pengambilan keputusan di mana setiap cabang dari pohon merepresentasikan keputusan, dan setiap daun mewakili hasil dari keputusan tersebut. Dalam konteks DecisionTreeClassifier, model ini digunakan untuk klasifikasi, di mana tujuannya adalah memisahkan data ke dalam kelas atau kategori yang berbeda. Keunggulan utama dari model Decision Tree adalah kemampuannya untuk melakukan parameter tuning yang cermat. Dengan menggunakan GridSearchCV, model dapat secara otomatis mencari kombinasi hyperparameter terbaik yang meningkatkan performa model dalam memprediksi data. Hasil dari parameter tuning membantu meningkatkan akurasi dan kinerja model secara keseluruhan, memungkinkan penggunaan model ini secara optimal dalam tugas klasifikasi. Dengan demikian, parameter tuning adalah langkah kunci dalam menyempurnakan model Decision Tree untuk mencapai hasil terbaik dalam pemrosesan data dan prediksi.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.tree import DecisionTreeClassifier # Import class DecisionTreeClassifier dari modul sklearn.tree

#START CODE
model1 =  DecisionTreeClassifier() # Inisialisasi model DecisionTreeClassifier tanpa hyperparameter
params = {
    'criterion': ['gini', 'entropy'],  # Kriteria pemilihan fitur
    'max_depth': [None, 10, 20, 30],  # Maksimum kedalaman setiap pohon
    'min_samples_split': [2, 5, 10],  # Minimum sampel yang dibutuhkan untuk membagi node
    'min_samples_leaf': [1, 2, 4],  # Minimum sampel yang dibutuhkan di setiap leaf node
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV # Import class GridSearchCV dari modul sklearn.model_selection
grid_DT = GridSearchCV( # Inisialisasi objek GridSearchCV
             estimator= model1, # Model yang akan dioptimalkan
             param_grid= params,  # Daftar hyperparameter yang akan diuji
             scoring = 'accuracy', # Metrik evaluasi yang digunakan
             n_jobs =10, # Core cpu yang digunakan
             cv =10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid_DT.fit(X_train,y_train) # Melakukan optimisasi hyperparameter menggunakan data training
grid_DT.best_params_ # Mendapatkan hyperparameter terbaik dari hasil optimisasi
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid_DT.predict(X_test) # Melakukan prediksi menggunakan model yang telah ditest

print(classification_report(y_test,y_pred)) # Menampilkan classification report
print("")
print(confusion_matrix(y_test,y_pred)) # Menampilkan confusion matrix
print("")
print(accuracy_score(y_test, y_pred)) # Menampilkan nilai akurasi
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

Model kedua yang saya gunakan adalah Random Forest. Random Forest adalah sebuah model machine learning yang beroperasi dengan cara membuat beberapa pohon keputusan (decision trees) dan menggabungkan hasil prediksi dari setiap pohon untuk menghasilkan prediksi akhir. Pada setiap pohon keputusan, fitur-fitur yang digunakan untuk pemilihan keputusan dibuat secara acak, sehingga setiap pohon memiliki keberagaman dalam pembelajaran. Pada codingan yang diberikan, model Random Forest pertama kali dibuat tanpa melakukan penalaan terhadap hyperparameter, seperti jumlah pohon (n_estimators) dan kedalaman maksimum pohon (max_depth). Kemudian, dengan menggunakan GridSearchCV, kita mencari kombinasi terbaik dari hyperparameter untuk meningkatkan kinerja model. Random Forest diandalkan karena beberapa alasan, termasuk kemampuannya dalam mengatasi overfitting, ketahanannya terhadap noise, dan kemampuannya untuk menangani data dengan fitur yang banyak. Pemilihan fitur secara acak dan penggunaan beberapa pohon membuat model ini sangat efektif dalam berbagai jenis data dan tugas klasifikasi.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.ensemble import RandomForestClassifier # Import class RandomForestClassifier dari modul sklearn.tree

#START CODE
model2 = RandomForestClassifier() # Inisialisasi model RandomForestClassifier tanpa hyperparameter
params = {
    'n_estimators' : [100, 200, 300],  # Jumlah pohon dalam ensemble
    'max_depth': [3, 5, None], # Maksimum kedalaman setiap pohon
    'max_features' : ['sqrt', 'log2'], # Jumlah fitur yang dipertimbangkan untuk split
    'min_samples_split' : [2, 4, 6] # Minimum sampel yang dibutuhkan untuk membagi node
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV # Import class GridSearchCV dari modul sklearn.model_selection
grid_RF = GridSearchCV( # Inisialisasi objek GridSearchCV
             estimator= model2, # Model yang akan dioptimalkan
             param_grid= params, # Daftar hyperparameter yang akan diuji
             scoring = 'accuracy', # Metrik evaluasi yang digunakan
             cv =10, # Core cpu yang digunakan
             n_jobs =10 # core cpu yang digunakan # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid_RF.fit(X_train,y_train) # Melakukan optimisasi hyperparameter menggunakan data training
grid_RF.best_params_ # Mendapatkan hyperparameter terbaik dari hasil optimisasi
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid_RF.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

Model ketiga yang digunakan dalam kode di atas adalah Support Vector Machine (SVM) melalui class SVC dari scikit-learn. SVM adalah model pembelajaran mesin yang sangat berguna untuk klasifikasi dan regresi. Pada kode tersebut, objek SVM dibuat dengan parameter default, kemudian dilakukan penyetelan parameter menggunakan Grid Search Cross Validation (GridSearchCV) untuk mencari kombinasi terbaik dari 'C' (cost of misclassification), 'kernel' (jenis kernel), dan 'gamma' (koefisien kernel) guna meningkatkan kinerja model SVM. Proses ini dilakukan dengan GridSearchCV menggunakan metode cross-validation, yang memastikan keakuratan dan ketahanan model. Keunggulan class SVC terletak pada kemampuannya menangani pemisah linier yang kompleks dalam data, sehingga efektif untuk masalah yang melibatkan pemisah yang rumit. Selain itu, SVM mampu menangani data dengan dimensi yang tinggi, menjadikannya relevan untuk berbagai jenis dataset dengan karakteristik yang berbeda. Dengan kombinasi parameter yang tepat, class SVC dapat memberikan hasil klasifikasi atau regresi yang sangat baik dalam berbagai aplikasi.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.svm import SVC # Import class SVC dari modul sklearn.svm

#START CODE
model3 = SVC() # Inisialisasi model Support Vector Classification (SVC) tanpa hyperparameter
params = {
    'C': [0.1, 1, 10],  # Parameter penalti
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Jenis kernel
    'gamma': ['scale', 'auto', 0.1, 1],  # Koefisien kernel
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV # Import class GridSearchCV dari modul sklearn.model_selection
grid_SVC = GridSearchCV( # Inisialisasi objek GridSearchCV
             estimator= model3, # Model yang akan dioptimalkan
             param_grid= params, # Daftar hyperparameter yang akan diuji
             scoring = 'accuracy', # Metrik evaluasi yang digunakan
             n_jobs = 10, # Core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid_SVC.fit(X_train,y_train) # Melakukan optimisasi hyperparameter menggunakan data training
grid_SVC.best_params_ # Mendapatkan hyperparameter terbaik dari hasil optimisasi
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid_SVC.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```
